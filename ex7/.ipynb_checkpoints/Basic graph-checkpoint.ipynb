```python
# 서울의 기온 데이터 읽어오기
import csv # csv 파일을 import하는게 아니라 csv를 사용하기 위한 모듈을 import하는 것
f = open('seoul.csv')
data = csv.reader(f)

for row in data: # in을 쓰면, 데이터가 없으면 자동으로 for문에서 빠져나옴
    print(row)
```

```python
# 데이터 읽어와서 최고 기온 데이터 출력하기 - 헤더를 제거하고 최고 기온 출력
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data) # 헤더를 제거 (첫 번째 줄)

for row in data:
    print(row[-1]) # 맨 뒤 데이터 (최고기온) 출력
```

```python
# 데이터 리스트에 저장하기 - 최고 기온 데이터를 모두 result 리스트에 저장
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data:
    if row[-1] != '': # 최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가 
print(result)
```

```python
# 데이터의 개수 확인
print(len(result))
```

```python
# 최고 기온 데이터 꺾은선 그래프로 그리기
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = [] 

for row in data:
    if row[-1] != '': 
        result.append(float(row[-1])) 

plt.plot(result, 'r') # result 리스트에 저장된 값을 빨간색 그래프로 그리기 # 리스트가 result 하나밖에 없음 -> x좌표는 index값
plt.show() # 그래프 나타내기 
```

```python
# 그래프 크기 조절하기
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 2), dpi = 300) # 가로 10인치, 세로 2인치, dpi값은 300 
plt.plot(result, 'r')
plt.show()
```

```python
# 8월의 최고 기온 데이터 시각화하기
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = [] 

for row in data: 
    if row[-1] != '': # 최고 기온 값이 존재한다면 
        if row[0].split('-')[1] == '08': # 8월에 해당하는 값이라면 
            result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가 
             
import matplotlib.pyplot as plt
plt.figure(dpi = 300) 
plt.plot(result, 'hotpink') # result 리스트에 저장된 값을 hotpink 색으로 그리기 
plt.show() # 그래프 나타내기 
```

```python
# 특정 날짜의 최고 기온 데이터 시각화하기 
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = [] 

for row in data: 
    if row[-1] != '': 
        if row[0].split('-')[1] == '08' and row[0].split('-')[2] == '06': # 날짜가 8월 6일일 때 
            result.append(float(row[-1])) # 매년 8월 6일 최고기온만 result에 들어감 
             
import matplotlib.pyplot as plt
plt.figure(dpi = 300) 
plt.plot(result, 'hotpink') 
plt.show() 
```

```python
# 특정 날짜의 최고 기온 및 최저 기온 데이터 시각화하기
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data: 
    if row[-1] != '' and row [-2] != '': # 최고기온과 최저기온이 존재하면 
        if 2001 <= int(row[0].split('-')[0]): # 연도가 2001 이상이면 
            if row[0].split('-')[1] == '08' and row[0].split('-')[2] == '06': # 날짜가 8월 6일일 때 
                high.append(float(row[-1])) # 최고기온을 high에 넣기
                low.append(float(row[-2])) # 최저기온을 low에 넣기 
             
import matplotlib.pyplot as plt
plt.figure(dpi = 300) 
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')
plt.show() 
```

```python
# 옵션 설정 
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = [] # 최고 기온 값을 저장할 리스트 high 생성
low = [] # 최저 기온 값을 저장할 리스트 low 생성 

for row in data: 
    if row[-1] != '' and row [-2] != '': # 최고기온과 최저기온이 존재하면 
        date = row[0].split('-') # 날짜 값을 - 문자를 기준으로 구분하여 저장 
        if 2001 <= int(date[0]): # 2001년 이후 데이터라면 
            if date[1] == '08' and date[2] == '06': # 날짜가 8월 6일이라면 
                high.append(float(row[-1])) # 최고 기온 값을 high 리스트에 저장
                low.append(float(row[-2])) # 최저 기온 값을 low 리스트에 저장 
                
                plt.rc('font', family = 'Malgun Gothic') # 내 컴퓨터에 맑은 고딕이 존재하면 적용, 존재하지 않으면 기본 글꼴 적용 
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지. 깨지지 않는다면 굳이 쓰지 않아도 됨 
plt.title('8월 6일의 기온 변화 그래프')
             
plt.plot(high, 'hotpink', label = 'high') # high 리스트에 저장된 값을 hotpink 색으로 그리고 레이블 표시 
plt.plot(low, 'skyblue', label = 'low') # low 리스트에 저장된 값을 skyblue 색으로 그리고 레이블 표시 
plt.legend() # 범례 표시
plt.show() # 그래프 나타내기 
```
