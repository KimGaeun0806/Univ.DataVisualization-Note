```python
# 최고 기온 데이터 히스토그램으로 표현하기
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1])) 
        
plt.hist(result, bins = 100, color = 'r') # 히스토그램으로 나타내기 # bins=100 -> 전체 범위를 100으로 나눈다는 뜻 
plt.show()
```

```python
# 8월의 최고 기온 데이터 히스토그램으로 표현하기 
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = [] # 8월의 최고 기온 값을 저장할 aug 리스트 생성 

for row in data:
    month = row[0].split('-')[1] # -로 구분된 값 중 2번째 값을 month에 저장
    if row[-1] != '':
        if month == '08': # 8월달이라면
            aug.append(float(row[-1])) # aug 리스트에 최고 기온 값 추가 
        
plt.hist(aug, bins = 100, color = 'r') 
plt.show()
```

```python
# 1월과 8월의 최고 기온 데이터 히스토그램으로 표현하기 
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = [] 
jan = []

for row in data:
    month = row[0].split('-')[1] 
    if row[-1] != '':
        if month == '08': 
            aug.append(float(row[-1])) 
        if month == '01':
            jan.append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```

```python
# 1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []] # 월별 데이터를 저장할 리스트 month 생성 (12개)

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1]) - 1].append(float(row[-1])) # 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 -> month[0])
        
plt.figure(figsize = (10, 5), dpi = 300)
plt.boxplot(month)
plt.show()
```

```python
# 8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([]) # day에 31개의 리스트 넣기. 빈 리스트를 31번 append

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2]) - 1].append(float(row[-1]))
        
plt.boxplot(day, showfliers = False) # showfliers=false -> 이상한 값 (noise)을 표시하지 x. 이 값을 제거하지 않으면 원이 생김 
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = [[] for i in range(31)]

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2]) - 1].append(float(row[-1]))
        
plt.style.use('ggplot') # 스타일을 줌 
plt.figure(figsize = (10, 5), dpi = 300)
plt.boxplot(day, showfliers = False) 
plt.show()
```
