```python
# plt.plot (데이터셋)
import matplotlib.pyplot as plt # matplotlib 라이브러리의 pyplot 모듈 불러온 후 'plt'라는 별명으로 부르기 (as, alias)
plt.plot([10, 20, 30, 40]) # 이렇게 리스트를 하나만 쓰게 되면 y축. y축만 제공될 때는 x축은 index값이 됨 
plt.show()
```

```python
# plt.plot(x축 데이터셋, y축 데이터셋)
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [12, 43, 25, 15]) # (x축, y축)
plt.show()
```

```python
# 그래프에 제목 넣기
import matplotlib.pyplot as plt
plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
# 그래프에 범례 넣기 
import matplotlib.pyplot as plt
plt.title('legent') 
plt.plot([10, 20, 30, 40], label = 'asc') # 증가를 의미하는 asc 범례
plt.plot([40, 30, 20, 10], label = 'desc') # 감소를 의미하는 desc 범례
plt.legend() # 범례 표시 
plt.show()
```

```python
# 그래프 색상 바꾸기
import matplotlib.pyplot as plt
plt.title('color') 
plt.plot([10, 20, 30, 40], color = 'skyblue', label = 'skyblue')
plt.plot([40, 30, 20, 10], 'pink', label = 'pink') # color='pink'가 아닌 'pink'로만 써도 괜찮음 
plt.legend() 
plt.show()
```

```python
# 그래프 선 스타일 바꾸기
import matplotlib.pyplot as plt
plt.title('linestyle')
plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed') # red dashed 그래프 
plt.plot([40, 30, 20, 10], color = 'g', ls = ':', label = 'dotted') # green dotted 그래프
plt.legend()
plt.show()
```

```python
# 그래프 선 모양 바꾸기
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label = 'circle') # 빨간색 원형 마커 그래프 # r. -> red circle
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up') # 초록색 삼각형 마커 그래프 # g^ -> green triangle
plt.legend()
plt.show()
```
